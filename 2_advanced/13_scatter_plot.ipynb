```python
# co relate between 2 attribute to get the insights 
import pandas as pd 
from datasets import load_dataset 
import matplotlib.pyplot as plt
import ast
# load data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

df
```

```python
# how median salary average co relates with job skills - bascially skill define the salary (insight)
# co relation between multiple variables 

df = df[df['job_title_short'] == 'Data Analyst']
df
```

```python
df_exploded = df.explode('job_skills')

skills_stat = df_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)

skills_stat.sort_values(by ='skill_count', ascending = False).head(5)
```

```python
skills_stat.plot(kind ='scatter', x ='skill_count', y ='median_salary')
```
